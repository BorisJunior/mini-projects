```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
# Chargement du fichier CSV
ecommerce_data = pd.read_csv('data.csv', encoding='ISO-8859-1', parse_dates=['InvoiceDate'])

# Affichage des premières lignes
ecommerce_data_head = ecommerce_data.head()
ecommerce_data_head
```

```python
#Nettoyage et prétraitement des données

# Suppression des entrées avec des valeurs manquantes
ecommerce_data_cleaned = ecommerce_data.dropna(subset=['Description', 'CustomerID'])

# Filtrage des entrées non pertinentes (annulations ou quantités négatives)
ecommerce_data_cleaned = ecommerce_data_cleaned[ecommerce_data_cleaned['Quantity'] > 0]
ecommerce_data_cleaned = ecommerce_data_cleaned[ecommerce_data_cleaned['CustomerID'] != 12346 ]

# Création d'une nouvelle colonne pour le total des ventes par transaction
ecommerce_data_cleaned['TotalSales'] = ecommerce_data_cleaned['Quantity'] * ecommerce_data_cleaned['UnitPrice']
ecommerce_data_cleaned = ecommerce_data_cleaned[ecommerce_data_cleaned['TotalSales'] != 0]


ecommerce_data_cleaned.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration des visualisations
sns.set_style('whitegrid')
plt.figure(figsize=(15, 5))

# Distribution de la Quantité
plt.subplot(1, 3, 1)
sns.boxplot(ecommerce_data_cleaned['Quantity'])
plt.title('Distribution de la Quantité')

# Distribution du Prix Unitaire
plt.subplot(1, 3, 2)
sns.boxplot(ecommerce_data_cleaned['UnitPrice'])
plt.title('Distribution du Prix Unitaire')

# Distribution des Ventes Totales
plt.subplot(1, 3, 3)
sns.boxplot(ecommerce_data_cleaned['TotalSales'])
plt.title('Distribution des Ventes Totales')

plt.tight_layout()
plt.show()
```

```python
def remove_outliers_using_iqr(df, column):
    """
    On va retitrer les valeurs aberrantes en utilisant la méthode IQR
    """
    # Calcul IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Definition des limites
    lower_bound = Q1 - 0.5 * IQR
    upper_bound = Q3 + 0.5 * IQR
    
    # Filter the data
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers for Quantity and UnitPrice
data_cleaned = remove_outliers_using_iqr(ecommerce_data_cleaned, 'Quantity')
data_cleaned = remove_outliers_using_iqr(ecommerce_data_cleaned, 'UnitPrice')
data_cleaned = remove_outliers_using_iqr(ecommerce_data_cleaned, 'TotalSales')

data_cleaned.shape
```

```python
# Extraction du mois et de l'année de la colonne 'InvoiceDate'
ecommerce_data_cleaned['YearMonth'] = ecommerce_data_cleaned['InvoiceDate'].dt.to_period('M')

# Agrégation des données pour obtenir les ventes totales par mois
monthly_sales = ecommerce_data_cleaned.groupby('YearMonth')['TotalSales'].sum()

# Visualisation des ventes mensuelles
plt.figure(figsize=(15, 6))
monthly_sales.plot()
plt.title('Tendances mensuelles des ventes')
plt.xlabel('Mois')
plt.ylabel('Ventes totales')
plt.grid(True)

plt.show()
```


La visualisation montre les tendances mensuelles des ventes pour la période couverte par le jeu de données.

Observations clés :
1. **Croissance générale** : Il y a une tendance à la hausse des ventes au fil du temps, avec quelques fluctuations.
2. **Pic à la fin de l'année** : Il y a un pic notable des ventes à la fin de l'année, probablement en novembre ou décembre, ce qui pourrait être lié aux achats des fêtes de fin d'année.
3. **Fluctuations mensuelles** : Outre le pic à la fin de l'année, d'autres mois montrent également des augmentations ou des baisses des ventes, ce qui pourrait être dû à divers facteurs promotionnels, saisonniers ou autres.


```python
# Extraction du jour de la semaine de la colonne 'InvoiceDate'
ecommerce_data_cleaned['Weekday'] = ecommerce_data_cleaned['InvoiceDate'].dt.day_name()

# Ordre des jours de la semaine pour la visualisation
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Agrégation des données pour obtenir les ventes totales par jour de la semaine
weekday_sales = ecommerce_data_cleaned.groupby('Weekday')['TotalSales'].sum().reindex(order)

# Visualisation des ventes hebdomadaires
plt.figure(figsize=(12, 6))
weekday_sales.plot(kind='bar', color='skyblue')
plt.title('Tendances hebdomadaires des ventes')
plt.xlabel('Jour de la semaine')
plt.ylabel('Ventes totales')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

La visualisation montre les tendances des ventes en fonction du jour de la semaine.

Observations clés :

1. **Jours de semaine** : Les ventes sont nettement plus élevées les jours de semaine, avec un pic le jeudi.
2. **Weekend** : Il y a une baisse significative des ventes pendant le weekend, en particulier le samedi, où les ventes sont presque nulles. Il est possible que l'entreprise ne fonctionne pas ou ait une activité réduite le weekend.
3. **Jeudi** : Le jeudi semble être le jour le plus actif en termes de ventes.

Ces informations pourraient être utiles pour planifier des promotions, gérer les stocks ou allouer des ressources en fonction des jours de la semaine.


```python
# Extraction de l'heure de la colonne 'InvoiceDate'
ecommerce_data_cleaned['Hour'] = ecommerce_data_cleaned['InvoiceDate'].dt.hour

# Agrégation des données pour obtenir les ventes totales par heure
hourly_sales = ecommerce_data_cleaned.groupby('Hour')['TotalSales'].sum()

# Visualisation des ventes horaires
plt.figure(figsize=(12, 6))
hourly_sales.plot(kind='bar', color='lightcoral')
plt.title('Tendances horaires des ventes')
plt.xlabel('Heure de la journée')
plt.ylabel('Ventes totales')
plt.grid(axis='y')
plt.show()
```


La visualisation montre les tendances des ventes en fonction de l'heure de la journée.

Observations clés :

1. **Début de la journée** : Il y a une faible activité de vente tôt le matin.
2. **Pics** : Les ventes commencent à augmenter à partir de 7 heures du matin, atteignent un pic autour de midi, puis un autre pic en début d'après-midi.
3. **Fin de journée** : Après 15 heures, les ventes commencent à diminuer et sont presque nulles après 20 heures.



```python

ecommerce_data_cleaned = ecommerce_data_cleaned.dropna(subset=['CustomerID'])
ecommerce_data_cleaned['TotalSpent'] = ecommerce_data_cleaned['Quantity'] * ecommerce_data_cleaned['UnitPrice']

grouped_data = ecommerce_data_cleaned.groupby('CustomerID').agg({
    'TotalSpent': 'sum', 
    'InvoiceNo': 'nunique', 
    'Quantity': 'mean',
    'InvoiceDate': 'mean'
})

grouped_data.columns = ['TotalSpent', 'NumberOfPurchases', 'AvgItemsPerPurchase', 'LastPurchaseDate']
grouped_data['DaysSinceLastPurchase'] = (pd.to_datetime('now') - pd.to_datetime(grouped_data['LastPurchaseDate'])).dt.days
grouped_data = grouped_data.drop('LastPurchaseDate', axis=1)


grouped_data.head()

```

```python
from sklearn.preprocessing import StandardScaler

# Initialisation du StandardScaler
scaler = StandardScaler()

# Normalisation des données
scaled_data = scaler.fit_transform(grouped_data)

# Conversion en DataFrame pour une meilleure lisibilité
scaled_data_df = pd.DataFrame(scaled_data, columns=grouped_data.columns, index=grouped_data.index)

scaled_data_df.head()
```

```python
from sklearn.cluster import KMeans

# Calcul de l'inertie pour différentes valeurs de K
inertia = []
range_values = range(1, 15)

for k in range_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Tracé de la méthode du coude
plt.figure(figsize=(10,6))
plt.plot(range_values, inertia, marker='o', linestyle='--')
plt.title('Méthode du coude')
plt.xlabel('Nombre de clusters')
plt.ylabel('Inertie')
plt.grid(True)
plt.show()
```

```python
# Application de K-means avec K=5
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(scaled_data)
grouped_data['Cluster'] = clusters


```

```python
from sklearn.decomposition import PCA

# Application de PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

# Création d'un DataFrame avec les deux composantes principales
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pc_df['Cluster'] = clusters

# Visualisation des clusters
plt.figure(figsize=(10, 8))
colors = ['r', 'g', 'b']

for cluster, color in zip(pc_df['Cluster'].unique(), colors):
    indices_to_keep = pc_df['Cluster'] == cluster
    plt.scatter(pc_df.loc[indices_to_keep, 'PC1'],
                pc_df.loc[indices_to_keep, 'PC2'],
                c=color, s=50)

plt.title('Visualisation des clusters après PCA')
plt.xlabel('Première composante principale')
plt.ylabel('Deuxième composante principale')
plt.legend(pc_df['Cluster'].unique(), title='Clusters')
plt.grid(True)
plt.show()
```
